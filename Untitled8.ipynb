```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import seaborn as sl
```

```python
data=pd.read_csv("data.csv")
data
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
sl.displot(data["B_Salary"])
plt.show()
```

```python
ms=MinMaxScaler()
```

```python
ms.fit(data[["B_Salary"]])
```

```python
data["New_minmax"]=pd.DataFrame(ms.transform(data[["B_Salary"]]),columns=["X"])
```

```python
data.head()
```

```python
sl.displot(data["New_minmax"])
sl.plot()
```
